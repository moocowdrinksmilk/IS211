# Plotted Histograms to show the distribution of our data
We have plotted 4 histograms
1. The distribution of time taken to navigate the forum
2. The distribution of the number of misclicks while navigating the forum
3. The distribution of the time taken to reply to a thread
4. The distribution of the number of misclicks before a user replies to a thread



```
import warnings
warnings.filterwarnings("ignore")

fig, axes = plt.subplots(1, 2, figsize=(13,5))
fig.suptitle('Histograms for the time taken to navigate through the forum', y = 1.05, size = 15)

axes[0].set_title('Version A')
axes[1].set_title('Version B')



ax1=sns.distplot(df_control['time_taken_forum'], ax = axes[0], hist=True, kde=True, 
             bins=int(10), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

ax1.set(xlabel='Time taken to navigate the forum (s)', ylabel='Density')

ax2 = sns.distplot(df_treatment['time_taken_forum'], ax = axes[1], hist=True, kde=True, 
             bins=int(10), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

ax2.set(xlabel='Time taken to navigate the forum (s)', ylabel='Density')


plt.show()
```

```
fig, axes = plt.subplots(1, 2, figsize=(13,5))
fig.suptitle('Histograms for the time taken to reply to a thread', y = 1.05, size = 15)

axes[0].set_title('Version A')
axes[1].set_title('Version B')


ax1 = sns.distplot(df_control['time_taken_reply'], ax = axes[0], hist=True, kde=True, 
             bins=int(15), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

ax1.set(xlabel='Time taken to reply to a thread (s)', ylabel='Density')

ax2 = sns.distplot(df_treatment['time_taken_reply'], ax = axes[1], hist=True, kde=True, 
             bins=int(15), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

ax2.set(xlabel='Time taken to reply to a thread (s)', ylabel='Density')

plt.show()
```

```
fig, axes = plt.subplots(1, 2, figsize=(13,5))
fig.suptitle('Histograms for number of misclicks while the user navigates the forum', y = 1.05, size = 15)

axes[0].set_title('Version A')
axes[1].set_title('Version B')


ax1 = sns.distplot(df_control['number_of_misclick_forum'], ax = axes[0], hist=True, kde=True, 
             bins=int(10), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

ax1.set(xlabel='Number of misclicks', ylabel='Density')

ax2 = sns.distplot(df_treatment['number_of_misclick_forum'], ax = axes[1], hist=True, kde=True, 
             bins=int(10), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

ax2.set(xlabel='Number of misclicks', ylabel='Density')

plt.show()
```

# Statistical Tests

<h5
>Below are the statistical test that we used using the SciPy library. We ustilised both the paired t-test and the Wilcoxon test to determine if there is a significant difference between the time taken and the number of misclicks between version A and B
</h5>


```
import scipy

scipy.stats.ttest_ind(df_control['time_taken_forum'], df_treatment['time_taken_forum'])
```

```
scipy.stats.ttest_ind(df_control['time_taken_reply'], df_treatment['time_taken_reply'])
```

```
scipy.stats.wilcoxon(df_control['number_of_misclick_forum'], df_treatment['number_of_misclick_forum'])
```

```
scipy.stats.wilcoxon(df_control['number_of_misclick_reply'], df_treatment['number_of_misclick_reply'])
```

```
scipy.stats.wilcoxon(df_control['number_of_misclick_reply'], df_treatment['number_of_misclick_reply'], alternative='less')
```

```
scipy.stats.wilcoxon(df_control['number_of_misclick_reply'], df_treatment['number_of_misclick_reply'], alternative='greater')
```
